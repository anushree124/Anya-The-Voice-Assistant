```python
import speech_recognition as sr
import webbrowser
import pyttsx3
import requests
from openai import OpenAI
from gtts import gTTS
import pygame
import os
import time

recognizer = sr.Recognizer()
engine = pyttsx3.init()

newsapi = "<Your Key Here>"
openai_api_key = "<Your Key Here>"

def speak(text):
    tts = gTTS(text)
    tts.save('temp.mp3')

    pygame.mixer.init()
    pygame.mixer.music.load('temp.mp3')
    pygame.mixer.music.play()

    while pygame.mixer.music.get_busy():
        pygame.time.Clock().tick(10)

    pygame.mixer.music.unload()
    os.remove("temp.mp3")

def aiProcess(command):
    client = OpenAI(api_key=openai_api_key)

    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a virtual assistant named Anya skilled in general tasks like Alexa and Google Cloud. Give short responses please"},
            {"role": "user", "content": command}
        ]
    )

    return completion.choices[0].message.content

def processCommand(command):
    if "exit" in command.lower() or "quit" in command.lower():
        speak("Goodbye!")
        return False
    elif "open google" in command.lower():
        webbrowser.open("https://google.com")
    elif "open facebook" in command.lower():
        webbrowser.open("https://facebook.com")
    elif "open twitter" in command.lower():
        webbrowser.open("https://twitter.com")
    elif "open youtube" in command.lower():
        webbrowser.open("https://youtube.com")
    elif "open linkedin" in command.lower():
        webbrowser.open("https://linkedin.com")
    elif "open instagram" in command.lower():
        webbrowser.open("https://instagram.com")
    elif "open google maps" in command.lower():
        webbrowser.open("https://maps.google.com")

    elif command.lower().startswith("play"):
        song = command.lower().split(" ", 1)[1]
        speak(f"Playing {song}")
        # Implement your music playing logic here
    elif "news" in command.lower():
        r = requests.get(f"https://newsapi.org/v2/top-headlines?country=in&apiKey={newsapi}")
        if r.status_code == 200:
            data = r.json()
            articles = data.get('articles', [])
            for article in articles[:3]:  # Limit to first 3 articles
                speak(article['title'])
    else:
        output = aiProcess(command)
        speak(output)
    return True

def listen_for_command(timeout=5, phrase_time_limit=5):
    with sr.Microphone() as source:
        print("Listening...üé§")
        recognizer.adjust_for_ambient_noise(source, duration=0.5)
        try:
            audio = recognizer.listen(source, timeout=timeout, phrase_time_limit=phrase_time_limit)
            command = recognizer.recognize_google(audio)
            print(f"Recognized: {command}")
            return command
        except sr.WaitTimeoutError:
            print("Listening timed out. No command detected.")
            return None
        except sr.UnknownValueError:
            print("Could not understand audio")
            return None
        except sr.RequestError as e:
            print(f"Could not request results; {e}")
            return None

def main():
    print("Initializing Anya...üöÄ")
    speak("Initializing Anya...")

    print("Hey there üëÄ, Welcome to Anya.")
    a = input("Enter your Name: ")
    print("Welcome", a)

    while True:
        try:
            print("Listening for commands...üé§")
            print(" Say 'Open Google' to open Google ‚Üó ")
            print(" Say 'Open Youtube' to open Youtube ‚Üó ")
            print(" Say 'Open LinkedIn' to open LinkedIn ‚Üó ")
            print(" Say 'Open Instagram' to open Instagram ‚Üó ")
            print(" Say 'Open Facebook' to open Facebook ‚Üó ")
            print(" Say 'Open Twitter' to open Twitter ‚Üó ")
            print(" Say 'Open Google Maps' to open Google Maps ‚Üó ")
            command = listen_for_command(timeout=5, phrase_time_limit=5)

            if command:
                print(f"Command recognized: {command}")
                continue_session = processCommand(command)

                if not continue_session:
                    break
                last_activity_time = time.time()
            elif time.time() - last_activity_time>20:
                    speak("Session timed out.")
                    print("Session timed out‚è±.")
                    print("See you next timeü•π")
                    print("We will miss you ü•π ", a )
                    exit()

            else:
                print("No command detected.Still listening.....")



            time.sleep(0.1)  # Small delay to prevent excessive CPU usage
        except Exception as e:
            print(f"An error occurred: {e}")
            exit()



    print("Anya has been terminated.")
if __name__ == "__main__":
    main()
```
